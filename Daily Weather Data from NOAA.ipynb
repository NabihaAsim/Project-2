#### Loop the API Call and append the DataFrame

```python
locationid = 'CLIM:101'

for i, row in dates_df.iterrows():
    date = row['date']
    url = 'https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid='+str(datasetid)+'&datatypeid='+str(datatypeid)+'&units='+str(units)+'&limit='+str(limit)+'&locationid='+str(locationid)+'&startdate='+str(date)+'&enddate='+str(date)
    api_response = requests.get(url, headers={'token':token})
    json_data = json.loads(api_response.text)
    data_df = pd.DataFrame(json_data['results'])
    data_df['climate_region'] = str(locationid)
    data_df.to_csv('data_export/weatherdata.csv', mode='a', header=False)
    print(f"Weather data for {locationid} and {date} added to weatherdata.csv file")
```

### Cleanup: Find and Delete Duplicate Rows


```python
# Import weatherdata.csv from folder data_export to dataframe weatherdata_df
weatherdata_df = pd.read_csv(r'data_export/weatherdata.csv', index_col=False, header=0)
weatherdata_df.head()
```

```python
weatherdata_df.count()
```

```python
# List duplicates to duplicates_df
duplicates_df = weatherdata_df[weatherdata_df.duplicated(['stationid','noaa_date'])] 
duplicates_df
```

```python
# Drop duplicates from weatherdata_df where both 'stationid' and 'noaa_date' are the same, keeping the first instance
weatherdata_df = weatherdata_df. drop_duplicates(subset=['stationid', 'noaa_date'], keep='first')
weatherdata_df.count()
```

```python
# Export dataframe to csv
weatherdata_df.to_csv('data_export/weatherdata.csv', index=False, header=True)
```
